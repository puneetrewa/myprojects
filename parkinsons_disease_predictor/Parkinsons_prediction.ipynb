```python
# Importing modules

import pandas as pd
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
#Load Dataset

parkinsons_data = pd.read_csv("parkinsons.csv") 
```

```python
#Set input and target variable 

X  = parkinsons_data.drop(['status','name'],axis = 1) 
y = parkinsons_data.iloc[:,-7]
```

```python
#now i create train and test set 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.266, random_state=21)
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
X_test
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:%.2f%%"% (accuracy*100))
```
